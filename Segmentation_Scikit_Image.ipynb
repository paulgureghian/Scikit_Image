# **Image Segmentation.** 

```
### Import an image
text = data.page()
plt.imshow(text) 
plt.grid(False) 
```

```
### Show the image pixel values and frequencies in a histogram 
fig, ax = plt.subplots(1, 1)
ax.hist(text.ravel(), bins=32, range=[0, 256])
ax.set_xlim(0, 256) 
```

```
### Supervised thresholding algorithm 
text_segmented = text > 50
plt.imshow(text_segmented) 
plt.grid(False)
```

```
### Unsupervised thresholding algorithms
### skimage 'otsu' method
text_threshold = filters.threshold_otsu(text) 
plt.imshow(text > text_threshold)
plt.grid(False) 
```

```
### skimage 'li' method
text_threshold = filters.threshold_li(text) 
plt.imshow(text > text_threshold)
plt.grid(False) 
```

```
### skimage 'local' method
text_threshold = filters.threshold_local(text, block_size=51, offset=10)  
plt.imshow(text > text_threshold)
plt.grid(False) 
```
